```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.info
```

```python
titanic.describe()
```

```python

# Basic pattern: survival by sex
sns.countplot(data=titanic, x='sex', hue='survived')
plt.title("Survival Count by Gender")
plt.show()
```

```python
titanic.loc[:,["survived","alive"]]
```

```python
titanic.iloc[:,[0,13]]
```

```python
# Boxplot: Age vs Sex, differentiated by Survival
sns.boxplot(data=titanic, x='sex', y='age', hue='survived')
plt.title("Boxplot of Age by Sex and Survival")
plt.show()
```

```python
# Violinplot: Age vs Sex, differentiated by Survival
sns.violinplot(data=titanic, x='sex', y='age', hue='survived')
plt.title("Violinplot of Age by Sex and Survival")
plt.show()
```

```python
# In a violin plot, the following symbols and elements are typically used:
# 1. Violin shape: Represents the kernel density estimation (KDE) of the data distribution. Wider parts indicate a higher density of data points.
# 2. White dot: Represents the median of the data within each category.
# 3. Thick black bar: Represents the interquartile range (IQR) or middle 50% of the data (from Q1 to Q3).
# 4. Thin black line: Represents the whiskers, which extend from the IQR to show the range of data (typically 1.5 * IQR from Q1 and Q3).
# 5. Outlier points: Individual points outside the whiskers, showing extreme values.
```
