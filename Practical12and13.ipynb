```python
# Get only the feature columns (exclude the 'species' column)
features = iris.columns[:-1]

# Create a figure with 4 rows and 2 columns of subplots
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(10, 10))

# Loop through each feature to plot histograms and boxplots
for i, feature in enumerate(features):
    # Plot histogram to see the distribution
    sns.histplot(iris[feature], ax=axes[i][0])
    axes[i][0].set_title(f'Distribution of {feature}')

    # Plot boxplot to detect outliers
    sns.boxplot(x=iris[feature], ax=axes[i][1])
    axes[i][1].set_title(f'Outliers in {feature}')

# Adjust layout to avoid overlap
plt.tight_layout()

# Show all plots
plt.show()
```
